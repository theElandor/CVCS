```python
import torch
```

```python
# plotting training loss of some checkpoints
path = "D:\\weights\\test\\balanced\\checkpoint50"
checkpoint = torch.load(path)
	
```

```python
TL = checkpoint['training_loss_values']
VL = checkpoint['validation_loss_values']
```

```python
import matplotlib.pyplot as plt
import numpy as np
group = 100
means = [np.mean(TL[i:i+group]) for i in range(0, len(TL), group)]
group=100
val_means = [np.mean(VL[i:i+group]) for i in range(0, len(VL), group)]
plt.plot(means)
```
